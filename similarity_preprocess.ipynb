```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('billboard_lyrics_2016-2017.csv').head(100)
```

```python
#make total words
df['Total Words'] = ""
for i in range(len(df)):
    total_words = 0
    if not(isinstance(df.iloc[i].Lyrics, float)): 
        lyrics = (df.iloc[i].Lyrics).split(' ')
        total_words = len(lyrics)
    df.loc[df.index[i], 'Total Words'] = total_words
    
for i in range(len(df)):
    print(df.loc[df.index[i], 'Total Words'])
```

```python
#make unique words
df['Unique Words'] = ""
for i in range(len(df)):
    unique_words = set()
    if not(isinstance(df.iloc[i].Lyrics, float)): 
        lyrics = (df.iloc[i].Lyrics).split(' ')
        for word in lyrics:
            unique_words.add(word)
    df.loc[df.index[i], 'Unique Words'] = len(unique_words)
    
for i in range(len(df)):
    print(int(df.loc[df.index[i], 'Unique Words']))
```

```python
overlap = np.zeros((len(df), len(df)))
for i in range(len(df)):
    for j in range(len(df)):
        if (i != j):
            sharedWords = set()
            if (isinstance(df.iloc[i].Lyrics, float) or isinstance(df.iloc[j].Lyrics, float)):
                overlap[i,j] = 0
                continue
            lyrics1 = (df.iloc[i].Lyrics).split(' ')
            lyrics2 = (df.iloc[j].Lyrics).split(' ')
            for word in lyrics1:
                if (word in lyrics2):
                    sharedWords.add(word)
            overlap[i,j] = len(sharedWords)

num_connections = 2
connections = set()
overlap_top = np.zeros((0,2))

for i in range(1, len(df)):
    rank = overlap[i].argsort()[-num_connections:][::-1]
    for j in range(len(rank)):
        if ((i, rank[j]) not in connections):
            overlap_top = np.vstack((overlap_top, [i, rank[j]]))
        connections.add((i, rank[j]))
        connections.add((rank[j], i))
        

np.savetxt("overlap.csv", np.vstack((np.arange(len(df)), overlap)), delimiter=",")
overlap_top = np.vstack((['Start', 'End'], overlap_top))
np.savetxt("overlap_top.csv", overlap_top, delimiter=",", fmt="%s")
```
